The first step is preprocessing <br>
Preparing data for feeding it to our learning model

```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
# loading the dataset
house_price_data = sklearn.datasets.load_boston()
house_price_data
```

```python
len(house_price_data)
```

```python
# for better structure we use pandas dataframe
house_price_dataset = pd.DataFrame(house_price_data.data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
        'TAX', 'PTRATIO', 'B', 'LSTAT'])

house_price_dataset.head()
```

```python
house_price_dataset.shape
```

As you can see, there is no "Price" column since we just feed .data to the DataFrame, and not the target

```python
# add the target column to the DataFrame

house_price_dataset["Price"] = house_price_data.target
house_price_dataset.head()
```

```python
# checing the dimensions of DataFrame

house_price_dataset.shape
```

```python
# Checking for missing values

house_price_dataset.isnull()
```

```python
house_price_dataset.isnull().sum()
```

There's no missing values

```python
# Statistical measures of the dataset

house_price_dataset.describe()
```

The very important thing in regressoin problems, is to understand the correlation between various features of the data

```python
# correlation of features

correlation = house_price_dataset.corr()
```

```python
# heatmap to visualize correlations

plt.figure(figsize=(10,8))

sns.heatmap(correlation, cbar=2, square=2, fmt=".1f", annot=True, annot_kws={"size":8}, cmap="YlGnBu")

# cbar : color bar (the bar on the right size ranging from minus to positive one)
# square : all the values show in a square
# fmt : how many floating values to be shown
# annot : annotation, writing features names in the left side
# cmap : color map
```

```python
# for our training we need to split features and the their targets

X = house_price_dataset.drop(["Price"], axis=1)  # if we are droping a column, the axis=1, if it is row, axis=0
y = house_price_dataset["Price"]
```

Now it is time to split the data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
print( "X_train shape : {}\n y_train shape : {}\n X_test shape : {}\n y_test shape : {}".format(X_train.shape, y_train.shape, X_test.shape, y_test.shape))
```

<h3> XGBoost Regressor </h3>
This is the model we are going to use as our regressor

```python
# loading the model

model = XGBRegressor()
```

```python
# training the model

model.fit(X_train, y_train)
```

Evaluating the model

```python
# making prediction 

# training predictions

train_data_predictions = model.predict(X_train)
train_data_predictions
```

```python
# R square error
score_R_square = metrics.r2_score(y_train, train_data_predictions)

# mean absolute error
score_mean_abs = metrics.mean_absolute_error(y_train, train_data_predictions)

print("R squared error : {}".format(score_R_square))
print("Mean Absolute error : {}".format(score_mean_abs))
```

Predictions on test data

```python
test_data_predictions = model.predict(X_test)
test_data_predictions
```

```python
# R square error
score_R_square = metrics.r2_score(y_test, test_data_predictions)

# mean absolute error
score_mean_abs = metrics.mean_absolute_error(y_test, test_data_predictions)

print("R squared error : {}".format(score_R_square))
print("Mean Absolute error : {}".format(score_mean_abs))
```
